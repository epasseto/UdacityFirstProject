```python
begin = time()

items_dic = {}
nan_count = 0
counter = 0

column = df_2020['LanguageWorkedWith']
for item in column:
    if udacourse.fn_isNaN(item):
        #print('NaN')
        nan_count += 1
    else:
        small_col = set(item.lower().split(';')) #I don´t want repeated items (if they exist!)
        for element in small_col:
            #print(element)
            if element in items_dic:
                items_dic[element] += 1
            else:
                items_dic[element] = 1
    if counter >= 5:
        break
    counter +=1

end = time()
print('elapsed time: {}s'.format(end-begin))
print('NaN total rows:', nan_count)
items_dic 
```
